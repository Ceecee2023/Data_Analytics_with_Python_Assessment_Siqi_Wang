## Shopping Cart System with Discounts

Write a Python program to simulate a shopping cart system for an online store. The program will calculate the total cost of items, apply discounts, and check if the total exceeds a specified budget.

1.	Variables and Lists:
  - Define a `budget` variable with an initial value of 200.
  
  - Create two empty lists called `item_names_list` and `item_prices_list` to store the name and price of each item separately.

```python
budget=200
item_name_list=[]
item_price_list=[]
```

2. Functions:
  - Write a function `add_item_to_cart(item_name, item_price)` that takes the itemâ€™s name and price as arguments, appends the name to item_names and the price to item_prices, and returns both updated lists.
  
  - Write a function `calculate_total(item_prices)` that calculates and returns the total cost of all items in item_prices.

    Conditions:
    - If the total cost exceeds the budget after adding an item, print "Budget exceeded!" and stop adding more items.
    - If the total cost is within budget and exceeds $100, apply a 10% discount on the total and print the discounted total.

```python
def calculate_total(item_prices):
    if sum(item_prices)>100:
        return sum(item_prices)*0.9
    else:
        return sum(item_prices)
```

```python
def add_item_to_cart(item_name, item_price):
    total_price = 0
    total_price = calculate_total(item_price_list)
    if total_price + item_price <= budget:
        item_name_list.append(item_name)
        item_price_list.append(item_price)
        print(f"Added {item_name} to cart")
    else:
        print("Budget exceeded!")
```

3.	Loop and Input:
  - Start the input only once the user says 'start'
  - Use a loop to allow the user to add items to the cart by entering an item name and price. The loop should stop when the user types 'done'.
  - For each item, add it to item_names and item_prices using add_item_to_cart, then update the total cost using calculate_total.

Output:
  - After the loop ends, display the final cart with each item and its price, the initial total, any applicable discount, and the final total.


```python
cost=0
user_input = input("Type 'start' to start shopping: ")
while user_input != "start":
    user_input = input("Type 'start' to start shopping: ")
while user_input != "done":
    item_name = input("Enter the item name: ")
    item_price = int(input("Enter the item price: "))
    add_item_to_cart(item_name, item_price)
    cost = calculate_total(item_price_list)
    user_input = input("Type 'done' to finish shopping: ")
    
print("Your cart includes:")
for i in range(len(item_name_list)):
    print(f"{item_name_list[i]}: ${item_price_list[i]}")
print(f"Initial total: ${sum(item_price_list)}")
if sum(item_price_list) > 100:
    print("10% discount applied")
print(f"Final total: ${cost}")
```

## Job Role and Workload Level Impact on Mental Health

Instructions: Investigate the influence of job roles and workload level on the mental health index.
- Create a new column `workload_level` that labels each entry as "High" if the `Hours_Worked_Per_Week` is above its mean, otherwise "Low." (5 marks)
- Group the dataset by `Industry` and calculate the average `Hours_Worked_Per_Week` for each combination. (5 marks)
- Use a bar chart to display the average `Stress_Level` for each job role, with separate bars for high and low stress levels. (5 marks)
- Analyze the results: Which job roles and workload levels appear to have the greatest impact on mental health? (5 marks)


```python
mean_hours_worked = data['Hours_Worked_Per_Week'].mean()
data['workload_level'] = data['Hours_Worked_Per_Week'].apply(lambda x: 'High' if x > mean_hours_worked else 'Low')

industry_avg_hours = data.groupby('Industry')['Hours_Worked_Per_Week'].mean()
print(industry_avg_hours)

stress_level_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
data['Stress_Level_Num'] = data['Stress_Level'].map(stress_level_mapping)

avg_stress_by_role_workload = data.groupby(['Job_Role', 'workload_level'])['Stress_Level_Num'].mean().unstack()

# Calculate the overall average stress level
overall_avg_stress = data['Stress_Level_Num'].mean()
avg_stress_by_role_workload['Average'] = overall_avg_stress

avg_stress_by_role_workload.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'salmon', 'lightgreen'], edgecolor='black')
plt.title('Average Stress Level by Job Role and Workload Level')
plt.xlabel('Job Role')
plt.ylabel('Average Stress Level')
plt.xticks(rotation=45)
plt.legend(title='Workload Level')
plt.show()

highest_stress_role_high = avg_stress_by_role_workload['High'].idxmax()
highest_stress_value_high = avg_stress_by_role_workload['High'].max()
highest_stress_role_low = avg_stress_by_role_workload['Low'].idxmax()
highest_stress_value_low = avg_stress_by_role_workload['Low'].max()

print(f"The job role with the highest average stress level under high workload is {highest_stress_role_high} with a stress level of {highest_stress_value_high}.")
print(f"The job role with the highest average stress level under low workload is {highest_stress_role_low} with a stress level of {highest_stress_value_low}.")
# Print stress level data by industry
stress_level_by_industry = data.groupby(['Industry', 'workload_level'])['Stress_Level_Num'].mean().unstack()
print(stress_level_by_industry)

# Statistics for Stress Levels by Industry and Workload Level
stress_level_stats_by_industry = data.groupby(['Industry', 'workload_level'])['Stress_Level_Num'].agg(['count', 'mean', 'std', 'min', 'max'])
print(stress_level_stats_by_industry)
```
